1. Q. What is Simple Linear Regression (SLR)? Explain its purpose.

>> A. Simple Linear Regression (SLR) is a statistical method used to model the relationship between two continuous variables: one independent variable (predictor, denoted as X) and one dependent variable (response, denoted as Y). It assumes that the relationship between these variables can be represented by a straight line.

The mathematical equation of SLR is:

Y = β₀ + β₁X + ε

Where:

> Y = Dependent variable (the outcome we want to predict)

> X = Independent variable (the predictor)

>β₀ = Y-intercept (the value of Y when X =0)

> β₁ = Slope (the change in Y for a one-unit change is X)

>  ε = Error term (random variation or residual)

The goal of SLR is to find the "best-fitting" straight line (called the regression line) through the data points by minimizing the sum of squared residuals (difference between observed and predicted Y values). This is typically done using the Ordinary Least Squares (OLS) method.

**Purpose of Simple Linear Regression:**

1. Prediction: To predict the value of the dependent variable based on the value of the independent variable (e.g.,predicting house price based on square footage).

2. Understanding relationships: To quantify the strength and direction of the linear relationship between two variables (e.g.,how strongly studying hours affect exam scores).

3. Inference: To test whether there is a significant linear relationship betweeen variables and estimate the effect size (slope). it is widely used in fields like economices, science, business, and social sciences for forcasting and decision-making when the relationship appear linear.



2. Q.  What are the key assumptions of Simple Linear Regression?

>> A. Simple Linear regression relies on several key assumptions for the results (estimates, predictions, and inference) to be valid and relable. Violating these assumptions can lead to biased or inefficient results. The main assumptions are:

1. Linearity: The relationship between the independent variable (X) and the dependent variable(Y) is linear. This means the mean of Y changes at a constant rate with X, forming a straight line.

2. Independence: The observations( data points) are independent of each other. The value of Y for one observation should not be influenced by the value of Y for another observation (no autocorrelation, especially important in time series data).

3. Homoscedasticity ( Constant Variance): The variance of the residuals (errors)is constant across all values of the independent variable X. The spread of residuals should be roughly the same for low and high values of X (no funnel-shaped patterns in residual plots).

4. Normality of Residuals : The residuals (differences between observed and predicted Y values) should be approximately normally distributed. This assumption is particularly important for hypothesis testing and confidence intervals (especially with small sample sizes).

5. No perfect multicollinearity: Since SLR has only one predictor, this is automatically satsifed (multicollinearity is more relevent in multiple regression).

Additional notes:

> The independent variable X is assumed to be measured without error (fixed or controlled).

> There should be no influential outliers that disproportionately affect the regression line.



3.  Write the mathematical equation for a simple linear regression model and
explain each term.

>> A. The mathematical equation for a simple linear regression (SLR) model is:

Ŷ = b₀ + b₁X

(or in population form: Y = β₀ + β₁X + ε)

Where:

> Ŷ (or Y in the population model): The predicted (or observed) value of the dependent variable (response/ outcome variable).

> b₀( or β₀): The y-intercept. it represents the estimated value of Ŷ when X =0 . It is the point where the regression line crosses the y-axis.

> b₁(or  β₁): The slope coefficient. It represents the estimated change in Y for a one-unit increase in X. A positive b indicates a positive relationship; a negativeb indicates a negative relationship.

> X : The independent variable (predictor/ explanatory variable).

>ε( error term, only in the population model):
The random error or residual, representing the differnce between the observed Y and the predicted  Ŷ. It captures unexplained variation.

The goal is to estimate b0 and b1 from sample data to best fit the line.

5. What is the method of least squares in linear regression?

>>A. The method of least squares (also called Ordinary Least Squares or OLS) is the standard technique used to estimate the parameters (b₀ and b₁) of a linear regression model.
It works by finding the regression line that minimizes the sum of the squared residuals.
•  A residual (eᵢ) is the vertical difference between an observed data point (Yᵢ) and the predicted value on the line (Ŷᵢ): eᵢ = Yᵢ - Ŷᵢ.
•  Squaring the residuals (eᵢ²) ensures positive values and penalizes larger deviations more heavily.
•  The objective is to minimize the Sum of Squared Residuals (SSR): SSR = Σ(eᵢ²) = Σ(Yᵢ - Ŷᵢ)².
This method produces the “best-fitting” straight line because it minimizes the total vertical deviation (in a squared sense) between the data points and the line. It has desirable statistical properties (unbiased, efficient under certain assumptions) and is computationally straightforward.

6. What is Logistic Regression? How does it differ from Linear Regression?

>> A. Logistic regression is a statisitical method used for binary classification problems, where the dependent variable (outcome) is categorical with two possible values (e.g., 0/1 , yes/no, true/false). It predicts the probability that a given input belongs to a particular category (usually the "positive" class, coded as 1).

The model uses the logistic (sigmoid )function to transform the output of a linear equation into a probability value between 0 and 1.

P(Y=1) = 1 / (1 + e^{-(β₀ + β₁X)})

(or in log-odds form: log(P/(1-P)) = β₀ + β₁X)
Here, e is the base of the natural logarithm, β₀ is the intercept, and β₁ is the coefficient for the predictor X.

**Key differences from linear Regression:**

> Type of outcome variable- Linear regression predicts continuous outcomes (e.g., house, price, temperature). Logistic regression predicts probabilities for categorical (binary) outcomes (e.g., pass/ fail, spam/not spam).

> Model equation and output - Linear regression uses a straight line (Y= b0 + b1X), which can produce any real number (unbounded). Logistic regression applies the sigmoid function to the linear combination, bounding the output between 0 and 1 (interpretable as probability).

> Estimation method- Linear regression uses ordinary least squares (minimizing squared errors). Logistic regression uses maximum likelihood Estimation(MLE) to find parameters that maximize the likelihood of observing the data.

> Use case- Linear regression is for regression(prediction of quantities). Logistic regression is for classification (though it outputs probabilities).

7. Q. Name and briefly describe three common evaluation metrics for regression
models.

>> A. Three common evaluation metrics for linear regression models (which predict continuous outcomes) are:
1.  Mean Squared Error (MSE): The average of the squared differences between predicted and actual values: MSE = (1/n) Σ(Yᵢ - Ŷᵢ)². It penalizes larger errors more heavily due to squaring and is in squared units of the target variable. Lower MSE indicates better fit.
2.  Root Mean Squared Error (RMSE): The square root of MSE: RMSE = √MSE. It is in the same units as the target variable, making it more interpretable than MSE. Lower RMSE is better; it measures average prediction error magnitude.
3.  Mean Absolute Error (MAE): The average of the absolute differences between predicted and actual values: MAE = (1/n) Σ|Yᵢ - Ŷᵢ|. It is robust to outliers (unlike MSE/RMSE) and also in the same units as the target. Lower MAE indicates better performance.
(Note: These metrics are not used for logistic regression, which uses metrics like accuracy, precision, recall, or log-loss.)

8. Q. What is the purpose of the R-squared metric in regression analysis?

>> A. The R-squared(R²) metric , also known as the coefficient of  determination, measures the proportion of the variance in the dependent variable(target) that is explained or predicted by the independent variable (s) in the regression model.
Its main purpose is to evaluate the goodness of fit of the model:

> It indicates how well the regression line fits the observed data points.

> Values range from 0 to 1 (or sometimes negative for poor models):

> R² = 1: The model perfectly explains all variance ( all points lie exactly on the regression line).

> R² = 0 : The model explains none of the variance ( no better than predicting the mean).

> R² close to 1 : Strong fit (high explanatory power).

> R² close to 0 : Weak fit.

it helps compare models and assess predictive strength, but it doesn't indicate causation or whether the model is correctly specified.

9. Q. Write Python code to fit a simple linear regression model using scikit-learn
and print the slope and intercept.
(Include your Python code and output in the code box below.)

>> A.

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Sample data ( reshape X to 20 array as required by scikitlearn)
X = np.array([[1], [2], [3], [4], [5]]) # Independent variable
y = np.array([1.2, 1.8, 3.0, 3.9, 5.1]) # Dependent variable

# Create a linear regression model
model = LinearRegression()
model.fit(X, y)

# Get and print slope (coefficient) and intercept
slope = model.coef_[0]
intercept = model.intercept_

print(f"Slope (Coefficient): {slope}")
print(f"Intercept: {intercept}")
```

10. Q.  How do you interpret the coefficients in a simple linear regression model?

>> A. In a simpel linear regression model, the equation is typically written as:

ŷ = β₀ + β₁x

(where ŷ is the predicted value of the dependent variable, X is the independent variable,  β₀ is the intercept, and  β₁ is the slope coefficient).

> The intercept(b0): This is the expected value of the dependent variable (y) when the independent variable (x) is equal to 0. it represents the baseline value or starting point on the y-axis where the regression line crosses when x = 0.

> The slope(b1): This represents the average change in the dependent variable (Y) for a one-unit increase in the independent variable (x).

- If b1 > 0 : Positive relationship (as x increase, y increases).

- if b1 < 0 : Negative relationship ( as x increases, Y decreases).

- The magnitude tells how strong the change is (e.g., b1 =2 means y increases by 2 units for every 1 unit increase in x).

These interpretations assume the model is correctly specified and hold in the context of the data's units.
